```python
import github
import requests
import urllib.request
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
urlinit = "https://gist.github.com/paulmillr/2657075"
```

```python
def get_soup_from_url(url):
    res = requests.get(url, verify=True, timeout = 4)
    soup = BeautifulSoup(res.text, 'html.parser')
    return soup
```

```python
def get_users(url):
    soup = get_soup_from_url(url)
    users=[]
    dd = soup.find('table').find('tbody').find_all('tr')
    for index, line in enumerate(dd):
         users.append(line.find('td').text.split(" ")[0])
    return users
```

```python
users=get_users(urlinit)
users
```

```python
def scrap_avg_star(user):
    tok = {'Authorization': 'token 5dbdb418b810668245fee3e60c5fac0d6e316ab3'}
    url = f"https://api.github.com/users/{user}/repos"
    response = requests.get(url, headers=tok)
    data = response.json()
    sum_star = 0
    avg_star = 0
    for datak in data:
        sum_star += datak['stargazers_count']
    if len(data) == 0:
        return 0    
    avg_star = sum_star/len(data)
    return avg_star
```

```python
def scrap_top_users(url):
    top_users = []
    users = get_users(url) 
    for userk in users:  
        data_user = {'user': userk, 'avg_star' : scrap_avg_star(userk) }
        top_users.append(data_user)
    return top_users
```

```python
users = pd.DataFrame(scrap_top_users(urlinit))
```
